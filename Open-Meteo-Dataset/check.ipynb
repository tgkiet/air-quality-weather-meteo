```python
import os
import glob

# --- ĐỊNH NGHĨA CÁC ĐƯỜNG DẪN THƯ MỤC ---
aq_cams_dir = "hanoi_air_quality_from04Aug_cams_raw"
weather_meteo_dir = "hanoi_weathermeteo_from04Aug_data_raw" # Sửa lại tên thư mục cho đúng với ảnh
# ----------------------------------------

print("--- KIỂM TRA SỐ LƯỢNG FILE THỰC TẾ ---")

# 1. Kiểm tra thư mục Air Quality CAMS
if os.path.exists(aq_cams_dir):
    aq_files = glob.glob(os.path.join(aq_cams_dir, "*.csv"))
    print(f"Trong thư mục '{aq_cams_dir}':")
    print(f" -> Tìm thấy {len(aq_files)} file .csv")
else:
    print(f"Thư mục '{aq_cams_dir}' không tồn tại.")

# 2. Kiểm tra thư mục Weather Meteo
if os.path.exists(weather_meteo_dir):
    weather_files = glob.glob(os.path.join(weather_meteo_dir, "*.csv"))
    print(f"\nTrong thư mục '{weather_meteo_dir}':")
    print(f" -> Tìm thấy {len(weather_files)} file .csv")
else:
    print(f"Thư mục '{weather_meteo_dir}' không tồn tại.")

# 3. So sánh với số lượng trạm gốc
metadata_filename = "stations_metadata.csv"
if os.path.exists(metadata_filename):
    import pandas as pd
    df_metadata = pd.read_csv(metadata_filename)
    num_stations = len(df_metadata)
    print(f"\nSố lượng trạm trong file metadata gốc là: {num_stations}")
else:
    print(f"\nKhông tìm thấy file metadata.")
```
